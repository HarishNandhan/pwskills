Q1. Matplotlib is a visualization library in Python. It is built on NumPy arrays and designed to work with the broader SciPy stack.
We can visualize the data by using the following plots:
1. Line Graph
2. Stem plot
3. Bar chart
4. Histogram
5. Scatter Plot

```python
#Q2
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y,color='red')
plt.title("Scatter Plot Example")
plt.xlabel("X normally distributed values")
plt.ylabel("Y normally distributed values")
plt.grid()
```
